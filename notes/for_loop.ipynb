# FizzBuzz
## Write a program that automatically prints the solution for the FizzBuzz game
1 The program should print each number from 1 to 100 (including 100)
2 All numbers divisible by 3 should be replaced by the word "Fizz"
3 All numbers divisible by 5 should be replaced by the word "Buzz"
4 All numbers divisible by 3 and 5 should be replaced by the word "FizzBuzz"



```python
for number in range(1,101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```
